```python
from Bandit import Bandit, EpsilonGreedy, ThompsonSampling, Visualization, comparison
```

## Needed Variables

```python
Bandit_Reward = [1, 2, 3, 4]
NumberOfTrials = 20000
```

## EpsilonGreedy

```python
myEpsilon = EpsilonGreedy(Bandit).experiment(Bandit_Reward, NumberOfTrials)
```

```python
EpsilonGreedy(Bandit).report(NumberOfTrials, myEpsilon, algorithm = "EpsilonGreedy")
```

## Visual EpsilonGreedy

```python
Visualization().plot1(NumberOfTrials, myEpsilon, 'Epsilon Greedy')
```

### ThomsponSampling

```python
myThompson = ThompsonSampling(Bandit).experiment(Bandit_Reward, NumberOfTrials)
```

```python
ThompsonSampling(Bandit).report(NumberOfTrials, myThompson, 'ThompsonSampling')
```

## Visualizing ThompsonSampling

```python
Visualization().plot1( NumberOfTrials,myThompson, algorithm ='Thompson Sampling')
```

## Compare EpsilonGreedy and ThompsonSampling

```python
Visualization().plot2(myEpsilon, myThompson)
```

```python
comparison(NumberOfTrials, myEpsilon, myEpsilon)
```
